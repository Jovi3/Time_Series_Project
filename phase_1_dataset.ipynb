### So ETF's don't exist for a single stock, but for a stocks bundled together, and then follow their movement. For example, if we want to look at apple and google, we can either look at them separately, or take an ETF that includes these stocks. Below, I have used ETF's that contain the top companiess, by market cap, in 2021 (asked GPT, need to reconfirm this)

```python
# Top-5 stocks in 2021
stocks = ["AAPL", "MSFT", "AMZN", "GOOGL", "TSLA"]

# ETFs containing these top stocks
etfs = ["SPY", "QQQ", "VTI", "IVV", "ARKK"]

start = "2021-01-04"
end = "2023-12-29"

stock_data = yf.download(stocks, start=start, end=end)
etf_data = yf.download(etfs, start=start, end=end)

print(stock_data.head())
print(etf_data.head())
```

### Quants predict, market returns, not prices, since prices don't have much meaning without underlying knowledge, so we want to predict, returns. Returns would help us recognise movements, trends etc 

### To calculate reutnrs, log returns etc, we only need close, so now creating a subset 

```python
import pandas as pd
import numpy as np

# ================================================
# 1. EXTRACT CLOSE PRICES FROM EXISTING DATA
# ================================================

# Index prices (use Close)
sp500_close = sp500["Close"]["^GSPC"].rename("SP500")
nasdaq_close = nasdaq["Close"]["^IXIC"].rename("NASDAQ")

# ETF prices (also Close)
etf_closes = etf_data["Close"]   # wide dataframe: SPY, QQQ, VTI, IVV, ARKK

# Combine into one price DataFrame
prices = pd.concat([sp500_close, nasdaq_close, etf_closes], axis=1)
prices = prices.dropna()


# ================================================
# 2. COMPUTE RETURNS + LOG RETURNS
# ================================================

returns = prices.pct_change().dropna()
log_returns = np.log(prices / prices.shift(1)).dropna()


# ================================================
# 3. COMPUTE 20-DAY ROLLING VOLATILITY
# ================================================

vol20 = returns.rolling(20).std()
vol20.columns = [c + "_vol20" for c in vol20.columns]


# ================================================
# 4. COMPUTE DIRECTION (UP = 1, DOWN = 0)
# ================================================

direction = (returns > 0).astype(int)
direction.columns = [c + "_direction" for c in direction.columns]


# ================================================
# 5. COMPUTE ALPHA (RETURNS vs SP500 AND NASDAQ)
# ================================================

# Alpha relative to SP500
alpha_sp500 = returns.subtract(returns["SP500"], axis=0)
alpha_sp500.columns = [c + "_alpha_sp500" for c in alpha_sp500.columns]

# Alpha relative to NASDAQ
alpha_nasdaq = returns.subtract(returns["NASDAQ"], axis=0)
alpha_nasdaq.columns = [c + "_alpha_nasdaq" for c in alpha_nasdaq.columns]


# ================================================
# 6. BUILD MASTER DATASET
# ================================================

master = pd.concat([
    prices.add_suffix("_price"),
    returns.add_suffix("_ret"),
    log_returns.add_suffix("_logret"),
    vol20,
    direction,
    alpha_sp500,
    alpha_nasdaq
], axis=1)

master = master.dropna()


# ================================================
# 7. PREVIEW + SAVE
# ================================================

print(master.head())
master.to_csv("market_features_master.csv")
print("Saved as market_features_master.csv")
```
